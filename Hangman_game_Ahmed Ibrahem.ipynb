## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to guess the fruit name correctly.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible in the output.
> ##### In this way, all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, the number of chances equals the word length + 2.
> ##### For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.
> ##### Make game repeats until player ends the game.
> ##### Bonus : Create a function that takes 3 parameters `Easy`, `Medium` and `Hard` where `Medium` has more difficult words to be guessed than `Easy` and so on.

```python
#1-Import random to get random data
import random
#2-create function for choose difficulty (easy,medium,hard)
def get_word(difficulty):
    words = {
        "easy": ["apple", "mango", "grape", "peach"],
        "medium": ["avocado", "pineapple", "pomegranate"],
        "hard": ["blackcurrant", "dragonfruit", "passionfruit"]}
#3-choose random word
    return random.choice(words[difficulty])
#4-create function to play the game
def play_game():
    print("Welcom to hangman game")
#5-create a loop to repeate the game
    while True:
        difficulty= input ("Choose difficulty (easy/medium/hard): ").lower().strip()
        if difficulty not in ["easy","medium","hard"]:
            print("Invalid choice.Try again")
            continue
#6-choose a secret word based on difficulty
        secret_word = get_word(difficulty)
        guessed_letters = set()
        displayed_word = ["_"] * len(secret_word)
        chances = len(secret_word) + 2
        print(f"The word has {len(secret_word)} letters.")
        print(f"You have {chances} chances.")
#7-create a loop to get user input and check the letter
        while chances > 0 and "_" in displayed_word:
            print("Word:", " ".join(displayed_word))
            print("Guessed letters:", " ".join(sorted(guessed_letters)))

            guess = input("Guess a letter: ").lower()    
            if len(guess) != 1 or not guess.isalpha():
                print("Please enter a single letter.")
                continue
#8-check if letter already guessed
            if guess in guessed_letters:
                print("You already guessed that letter.")
                continue       
            guessed_letters.add(guess)
            if guess in secret_word:
                print("Correct guess!")
                for i, letter in enumerate(secret_word):
                    if letter == guess:
                        displayed_word[i] = guess
            else:
                chances -= 1
                print(f"Wrong guess! Chances left: {chances}")
#9-check if the user won or lost
        if "_" not in displayed_word:
            print("Congratulations! You guessed the word:", secret_word)
        else:
            print("Game Over! The word was:", secret_word)
#10-ask the user if they want to play again
        again = input("Do you want to play again? (y/n): ").lower()
        if again != "y":
            print("Thanks for playing!")
            break
```

```python
play_game()
```
